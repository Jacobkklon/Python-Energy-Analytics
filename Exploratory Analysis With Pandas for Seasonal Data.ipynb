```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
```

```python
#Loading in data:
solar_df = pd.read_excel('US_Solar_2019.xlsx')
solar_df.tail()
```

```python
#Question 1: Calculating Means and Modes of operating years 
op_df = solar_df['Operating Year']
op_df.median()
```

```python
op_df.mode(dropna=True) #This tells us that 2017 is the mode 
```

```python
#Question 2:
y = op_df.sort_values()
x = range(len(y))
plt.scatter(x,y)
```

```python
#Median of operating year can be interpreted as the point where 50% of the data is beneath it. This is roughly determined
#by seeing how high on the graph (to what year) 50% of the data is.
#Mode of operating year can be interpreted as the point that is the "most flat", which would imply the most repeats
# of a particular measurement
```

```python
#Question 3:
plt.hist(y)
#On this histogram, the x axis is year and the y axis is the count per year. It should be noted that the bins
#are multiple years
plt.xlabel('Operating Year')
plt.ylabel('Frequency of Measurement')
```

```python
#Question 4:
#Nameplate capacity is the maximum rated output of a powerplant (in this case) under typical conditions
```

```python
#Question 5:
nameplate_df = solar_df['Nameplate Capacity (MW)']
nameplate_df.mean() #This tells us the mean nameplate capacity in MW
```

```python
nameplate_df.var() #This tells us the variance of the nameplate capacities in (MW)^2
```

```python
#Question 6:
solar_df = solar_df.drop(['Utility ID','Operating Year','Plant Code'],axis='columns')
new_df = solar_df.groupby('State') #Grouping columns by state
new_df.mean() #Showing the mean solar capacities of each state
```

```python
new_df.sum() #This shows the sum of solar capacities for each state
```

```python
#Question 7:
mean_df = new_df.mean()
total_df = new_df.sum()
mean_df = mean_df.sort_values('Nameplate Capacity (MW)') #Sorting states by mean nameplate capacity
```

```python
total_df = total_df.sort_values('Nameplate Capacity (MW)') #Sorting states by total nameplate capacity
```

```python
#Creating two bar charts showing mean and total capacities between states. This highlights which states have the most 
#total capacity and which states generally have installations with larger solar capacities
plt.figure(figsize=(15,10))
plt.bar(mean_df.index,mean_df['Nameplate Capacity (MW)'])
plt.title('Comparison of Mean Nameplate Capacities')
plt.xlabel('State')
plt.ylabel('Mean Nameplate Capacity (MW)')
```

```python
plt.figure(figsize=(15,10))
plt.bar(total_df.index,total_df['Nameplate Capacity (MW)'])
plt.title('Comparison of Total Nameplate Capacities')
plt.xlabel('State')
plt.ylabel('Total Nameplate Capacity (MW)')
```
